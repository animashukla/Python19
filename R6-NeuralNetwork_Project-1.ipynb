# 2. Distinguish the feature and target set (2.5 points)

```python
target_data = data2["Exited"]
feature_data = data2.iloc[:, :13]
```

```python
y= np.array(target_data.values.astype('float32'))
```

# 3. Divide the data set into training and test sets ( 2.5 points)

```python
from sklearn.model_selection import train_test_split
X = data2.drop('Exited',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=2)
```

# 4. Normalize the train and test data (5 points)

```python
from sklearn.preprocessing import StandardScaler
X_scaled= StandardScaler().fit_transform(X)
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=1)
```

# 5. Initialize & build the model (10 points)

```python
# Initialize Sequential model

def model_seq():
    model = tf.keras.models.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(30, input_dim=13, activation='relu'),
    tf.keras.layers.Dense(40, input_dim=30, activation='relu'),
    tf.keras.layers.Dense(1, input_dim=40, activation='sigmoid')
    ])
    
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
model2=model_seq()
```

```python
model2.fit(X_scaled_train, y_train, validation_data=(X_scaled_test, y_test), epochs=50, batch_size=32)
```

# 6. Predict the results using 0.5 as a threshold (5 points)

```python
y_pred=model2.predict(X_scaled_test)
```

```python
y_pred_df=pd.DataFrame(data=y_pred, columns=['Predict'])
```

```python
y_scaled_pred=y_pred_df.round()
```

```python
y_pred_df.loc[y_pred_df.Predict>=0.5,'Predict']=1
y_pred_df.loc[y_pred_df.Predict>=0.5,'Predict']=0
```

# 7. Print the Accuracy score and confusion matrix (2.5 points)

```python
from sklearn.metrics import accuracy_score
score = (accuracy_score(y_test, y_scaled_pred.values)*100)
```

```python
print('Accuracy score :: ', score)
```

```python
from sklearn.metrics import confusion_matrix
print("\n***Confusion matrix***\n",confusion_matrix(y_test,y_scaled_pred.values))
```
